```python
x_axis = [i * 2 for i in range(1, 500)]
y_axis = [i ** 2 for i in range(1, 500)]
z_axis = [5 * i for i in range(1, 500)]
len(z_axis)
```

```python
import matplotlib.pyplot as plt
NUM_COLORS = 6

cm = plt.get_cmap('gist_rainbow')
fig, ax = plt.subplots(figsize=(20,10))
ax.scatter(x_axis, y_axis, z_axis, c = [cm(1.*i/NUM_COLORS) for i in range(NUM_COLORS)])
```
